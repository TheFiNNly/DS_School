Посмотрим на пропуски

```python
def percent_missing(data, miss):
    percent_nan = 100 * data.isnull().sum() / len(data)
    percent_nan = percent_nan[percent_nan>miss].sort_values()
    return percent_nan
```

```python
percent_nan = percent_missing(data, 10) # посмотрим на те столбцы, где больше 10% пропусков
```

```python
percent_nan
```

Экспертное мнение подсказывает, что столбцы с пропусками более 40% значений вряд ли будут полезны. Однако, данные с "годом постройки" могут быть полезны, посмотрим на года постройки поближе

```python
data_year = data[['build_count_after_1995', 'build_count_1971-1995','build_count_1946-1970', 
                 'build_count_1921-1945', 'build_count_before_1920', 'build_year']]
```

```python
data_year.head()
```

Т.к. пропусков в годах постройки нашего целевого объекта больше, чем пропусков по району в целом, 
то мы можем заполнить пропуск тем значением, которое наиболее часто встречается в районе. Будет очень примерно, но хоть что-то

```python
set_year = {'build_count_after_1995': 2000, 'build_count_1971-1995': 1983, 'build_count_1946-1970': 1958,
           'build_count_1921-1945': 1933, 'build_count_before_1920': 1915}
```

```python
data_year_na = data_year[data_year['build_year'].isnull()]
data_year_na['build_year'] = data_year.idxmax(axis=1)
data_year_na['build_year'] = data_year_na['build_year'].map(set_year)
data_year['build_year'] = data_year['build_year'].fillna(data_year_na['build_year'])
```

```python
data_year.head()
```

Теперь проверим, какие у нас значения остались в годах постройки

```python
data_year.describe()
```

Как будто в максимальных и минимальных значениях года постройки что-то не то. Проверим и поправим.

```python
data_year.loc[(data_year['build_year'] > 2020)]
```

```python
data_year.loc[(data_year.build_year == 4965), 'build_year'] = 1958
data_year.loc[(data_year.build_year == 20052009), 'build_year'] = 1958
```

```python
data_year.loc[(data_year['build_year'] < 1000)]
```

С минимальными значениями чуть позже. Пока обработаем оставшиеся пропуски. К сожалению, больше зацепиться не за что в части годов постройки, поэтому просто заполним средним значением.

```python
data_year = data_year.fillna(data_year.mean())
```

```python
data_year.head()
```

Вернемся к неадекватно маленьким значениям года постройки. Применим тут такой же подход, как и к "пропущенным" значениям чуть ранее

```python
data_year.loc[(data_year['build_year'] < 1000)]
```

```python
data_year_min = data_year.loc[(data_year['build_year'] < 1000)]
data_year_min['build_year'] = data_year_min.idxmax(axis=1)
data_year_min['build_year'] = data_year_min['build_year'].map(set_year)
data_year['build_year'].loc[(data_year['build_year'] < 1000)] = data_year_min['build_year']
```

```python
data_year.isnull().sum()
```

```python
data_year = data_year.fillna(data_year.mean())
```

```python
data_year.isnull().sum()
```

```python
data_year.describe()
```

Общая площадь и жилая площадь важные показатели. Общая площадь без пропусков, а жилая с определенным количеством пропусков. Можем вывести коэф отношения средних общей площади и жилой, и применить этот коэф к пропущенным значениям.
Остальные показатели не столь критичны и там особо ничего не выдумать, поэтому их потом обработаем стандартным способом

```python
data_sq = data[['full_sq','life_sq']][data['life_sq'].notna()]
```

```python
data_sq.head()
```

```python
coef = data_sq['full_sq'].mean() / data_sq['life_sq'].mean()
coef
```

```python
data['life_sq'].isnull().sum()
```

```python
data['life_sq'] = data['life_sq'].fillna(data['full_sq'].mean() / coef)
```

```python
data['life_sq'].isnull().sum()
```
